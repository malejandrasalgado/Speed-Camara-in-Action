```python
# importing modules
import os
import glob
import pandas as pd
import datetime as dt
from datetime import date  
```

## Camera Data

```python
# importing the dataset by reading the log files
inputPath = 'C:/Users/aleja/OneDrive/Bootcamp/projects/Final Project/Data'
os.chdir(inputPath)
extension = 'log'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
# create an empty list
file_dfs = []
```

```python
# loop for reading the files
iter_len = len(all_filenames)
for i in range(iter_len):
    input_file = inputPath+"/"+all_filenames[i]
    # Append it to the master df
    try:
        print("Append ", input_file)
        csv_df = pd.read_csv(input_file, header=None)
        print(csv_df)
        file_dfs.append(csv_df)
#             break
    except:
        print(input_file, " not found")
        continue
```

```python
# Concat all of the input files into one master dataframe
master_df = pd.concat(file_dfs)
```

```python
# named the columns
master_df.columns = ['Date', 'Time', '2', '3', '4',
                     'Lane', 'Speed', '7', '8', '9', '10', '11', '12']
```

```python
# dropping columns
Camera = master_df.drop(
    columns=['2', '3', '4', '7', '8', '9', '10', '11', '12'])
```

```python
# pandas convert float to int
Camera['Time'] = Camera['Time'].astype(int)
```

```python
# Add leading zeros to the integer column in Python
Camera['Time'] = Camera['Time'].apply(lambda x: '{0:0>6}'.format(x))
# new_df['Time'] = pd.to_datetime(new_df['Time'].astype(str), format='%H:%M:%S')
Camera['Time'] = pd.to_datetime(Camera['Time'], format='%H%M%S').dt.time
```

```python
# Tidying up the date field in the data
Camera['Date'] = Camera['Date']+20000000
```

```python
# formatting the date to datatime
Camera['Date'] = pd.to_datetime(Camera['Date'].astype(str), format='%Y%m%d')
Camera['Hour'] = Camera['Time'].astype(str)
Camera['Hour'] = Camera['Hour'].apply(lambda x: x[:2])
```

## Weather Data

```python
# File to Load
file_to_load = "history_weather.csv"
```

```python
# Read history weather File and store into Pandas data frame
history_weather = pd.read_csv(file_to_load)
```

```python
# Dropping columns
weather = history_weather.drop(columns=['Name', 'Maximum Temperature', 'Minimum Temperature', 'Wind Chill', 'Heat Index', 'Snow',
                                        'Snow Depth', 'Wind Gust', 'Visibility', 'Cloud Cover', 'Relative Humidity'])
```

```python
# Tidying up the date time field in the data
weather['Date time'] = pd.to_datetime(weather['Date time'])
weather['Weather Date'] = pd.to_datetime(weather['Date time']).dt.date
weather['Weather Time'] = pd.to_datetime(weather['Date time']).dt.time
```

```python
# Dropping columns
weather.drop(columns=['Date time'])
weather = weather[['Weather Date', 'Weather Time',
                   'Temperature', 'Precipitation', 'Wind Speed', 'Wind Direction', 'Conditions']]
```

```python
# formatting the date to datatime
weather['Weather Date'] = pd.to_datetime(
    weather['Weather Date'], infer_datetime_format=True)
```

```python
# Extract hour
weather['Weather Hour'] = pd.to_datetime(
    weather['Weather Time'], format='%H:%M:%S').dt.hour

weather['Weather Hour'] = weather['Weather Hour'].astype(str)
weather['Weather Hour'] = weather['Weather Hour'].apply(lambda x: '{0:0>2}'.format(x))
weather['Weather Hour'] = weather['Weather Hour'].apply(lambda x: x[:2])
weather.rename(columns={"Weather Date": "Date",
                        "Weather Time": "Time", "Weather Hour": "Hour"},  inplace=True)
```
